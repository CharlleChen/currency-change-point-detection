```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/WS_XRU_D_csv_row.csv')
# df.apply(pd.to_numeric)
headers = df.iloc[1].values
df = df.iloc[10:, :]
df.columns = headers
df.set_index('Currency', inplace=True)
df.index.name = 'Trade Date'
df.head()
```

```python
currencies = ['AUD', 'NZD', 'JPY', 'SEK', 'GBP', 'EUR', 'CHF']

for cur in currencies:
    for col in df.columns:
        if cur in col:
            one = df[[col]].iloc[:, :1].dropna()
            one.columns = ['Settle']
            print(cur, one.index.min(), one.index.max(), len(one))
            one.to_csv(f'data/currency/{cur}.csv')
            break
            
```

```python
all_cur = df.copy()

for col in all_cur.columns:
    drop = True
    for cur in currencies:
        if cur in col:
            drop = False
            all_cur = all_cur.rename(columns={col:cur})
            break
    if drop:
        all_cur = all_cur.drop(columns=col)
            # one = df[[col]].iloc[:, :1].dropna()
            # one.columns = ['Settle']
            # print(cur, one.index.min(), one.index.max())
            # one.to_csv(f'data/currency/{cur}.csv')
            # break
all_cur = all_cur.loc[:, ~all_cur.columns.duplicated()].dropna()
print(all_cur.index.min(), all_cur.index.max(), len(all_cur))
all_cur.to_csv(f'data/currency/all_cur.csv')
```
